```python
import pandas as pd 

planilha = pd.read_excel("..//dados//Ordenação.xlsx")

#Ordenando pela coluna de Vendedor e pela coluna de Total de Vendas
#Se a primeira coluna for igual, a segunda coluna entra como segunda condição.
#por padrão a ordem é crescente. 
ordenarVendedor = planilha.sort_values(by = ["Vendedor","Total Vendas"]) 

display(ordenarVendedor)
```

```python
import pandas as pd 

planilha = pd.read_excel("..//dados//Ordenação.xlsx")

#Ordenando pela coluna de Vendedor e pela coluna de Total de Vendas
#Ascending - ordem crescente
#Ascending por padrão = True
ordenarVendedor = planilha.sort_values(by = ["Vendedor","Total Vendas"], ascending = False) 

display(ordenarVendedor)
```
