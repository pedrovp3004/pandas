```python
import pandas as pd
import numpy as np 

notasAlunos_DF = pd.DataFrame({

    "Nome" : ["Pedro", "João", "Bento"],
    "Av1" : [10, 7.5,3.5],
    "Av2" : [6.5, 9.5, 7.5],
    "Av3" : [10,8,9],
    "Av4" : [3.5, 2.0,5.0],
    
})

#Adicionando nova coluna e calculando a média 
notasAlunos_DF["Média"] = (notasAlunos_DF["Av1"] + notasAlunos_DF["Av2"] + notasAlunos_DF["Av3"] + notasAlunos_DF["Av4"]) / 4

#Adicionando uma coluna com uma dict
colFaltas = [0,3,2]
notasAlunos_DF["Faltas"] = colFaltas 

#Alterar a nota de um aluno 
notasAlunos_DF.loc[0,"Av2"] = 10

#Recalculando a coluna média 
notasAlunos_DF["Média"] = (notasAlunos_DF["Av1"] + notasAlunos_DF["Av2"] + notasAlunos_DF["Av3"] + notasAlunos_DF["Av4"]) / 4

display(notasAlunos_DF)
```
