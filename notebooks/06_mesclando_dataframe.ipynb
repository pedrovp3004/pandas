```python
import pandas as pd 

#Arquivo de Vendas do mês de Janeiro
planilha_vendas_jan = pd.read_excel("..//dados//Vendas_Janeiro.xlsx")

#Arquivo de Vendas do mês de Fevereiro 
planilha_vendas_fev = pd.read_excel("..//dados//Vendas_Fev.xlsx")


#Unindo os DataFrame do mês de Janeiro e Fevereiro com o pd.concat([df1, df2,...dfn])
planilha_vendas_jan_fev = pd.concat([planilha_vendas_jan, planilha_vendas_fev])

#Resumindo/Pegando apenas 3 colunas do DF que unimos.
resumindo = planilha_vendas_jan_fev[["Vendedor", "Data Venda", "Total Vendas"]]

display(resumindo)
```
