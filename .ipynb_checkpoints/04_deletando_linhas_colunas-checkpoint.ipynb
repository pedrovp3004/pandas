```python
import pandas as pd

tabela = pd.read_excel("dados//Deletar_Linhas_Colunas.xlsx")

#Dropna(): Deleta linhas que tem pelo menos um valor em branco
deletando_linhas_em_branco = tabela.dropna()

#Drop(columns = "col1", "col2"..."coln") - Deleta uma ou mais colunas
deletando_duas_colunas = deletando_linhas_em_branco.drop(columns = ["Produto", "Data Venda"])

#axis = eixo(0 - linha/ 1 = column). 
deletandoLinha3 = deletando_duas_colunas.drop(2,axis=0) 

#Deletando duas linhas.
deletandoDuasLinhas = deletandoLinha3.drop([0,1], axis = 0)

display(deletandoDuasLinhas) 
```
